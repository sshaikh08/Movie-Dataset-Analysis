# Analysis of The Movie Data Base Data
## The following analysis explores the various attributes of successful movies

### It should be noted that a movie's success can be defined a few different ways based on this data. Successfull movies and how one success may relate to another (e.g voter score vs. total revenue) will be explored by answering the following questions

### -What were the most expensive movies each and their adjusted budgets for all available years in the data, and how did they compare to the most profitable and highly voted movies for that year?

### -Which genres had the highest adjusted revenue year to year? Which genres were the most expensive to produce? How did the adjusted budgets of the most profitable movies compare to the most expensive genres year to year?

###  -What movies had the largest adjusted revenues every year? How did they compare when analysing and comparing their various attributes (e.g genre, voter score, runtime, budget, voting score)?

### -How does runtime effect adjusted revenues?

```python
def create_df_max_by_release_year(tmdb_dataframe, column_name_string):
    column_maxes_by_release_year = tmdb_dataframe.groupby('release_year')[column_name_string].max()
    max_movies_per_year = tmdb_dataframe.loc[tmdb_dataframe[column_name_string].isin(column_maxes_by_release_year.values)].sort_values(by=['release_year'])
    return max_movies_per_year
```
